```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
netflix_rating = pd.read_csv("netflix_ratings_data.csv")
```

```python
rating_time = pd.DataFrame([netflix_rating.type, netflix_rating.year, netflix_rating.averageRating]).transpose()
```

```python
movie_rating = rating_time[rating_time.type == "Movie"].groupby('year').averageRating.mean()
tvshow_rating = rating_time[rating_time.type == "TV Show"].groupby('year').averageRating.mean()
```

```python
sns.lineplot(x = movie_rating.index, y = movie_rating)
sns.lineplot(x = tvshow_rating.index, y = tvshow_rating)
plt.title(label = 'Movie and TV Show Ratings Over Time')
plt.legend(labels = ['Movies', 'TV Shows'])
plt.savefig('line_plot_ratings_over_time.png')
```
